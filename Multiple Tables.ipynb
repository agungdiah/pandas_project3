```python
# Join or Merge in Pandas
```

```python
# https://www.datasciencemadesimple.com/join-merge-data-frames-pandas-python/
```

```python
import pandas as pd
import numpy as np
 
# data frame 1
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6]),
  'Product':pd.Series(['Oven','Oven','Oven','Television','Television','Television'])}
df1 = pd.DataFrame(d1)
 
 
# data frame 2
d2 = {'Customer_id':pd.Series([2,4,6,7,8]),
    'State':pd.Series(['California','California','Texas','New York','Indiana'])}
df2 = pd.DataFrame(d2)
```

```python
df1 
```

```python
df2
```

```python
inner_join_df= pd.merge(df1, df2, on='Customer_id', how='inner')
inner_join_df 
```

```python
outer_join_df=pd.merge(df1, df2, on='Customer_id', how='outer')
outer_join_df
```

```python
left_join_df= pd.merge(df1, df2, on='Customer_id', how='left')
left_join_df 
```

```python
right_join_df= pd.merge(df1, df2, on='Customer_id', how='right')
right_join_df 
```

```python
# Concatenate DataFrames
```

```python
# https://www.statology.org/concatenate-two-pandas-dataframes/
```

```python
df1 = pd.DataFrame({'team': ['A', 'A', 'A', 'A'],
                    'assists': [5, 7, 7, 9],
                    'points': [11, 8, 10, 6]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'team': ['B', 'B', 'B', 'B'],
                    'assists': [4, 4, 3, 7],
                    'points': [14, 11, 7, 6]})
```

```python
df2
```

```python
df3 = pd.concat([df1, df2])
```

```python
df3
```

```python
#concatenate the DataFrames and ignore index
df3 = pd.concat([df1, df2], ignore_index=True)
```

```python
df3
```
