```python
import pandas as pd
```

```python
visits = pd.read_csv('visits.csv')
```

```python
visits
```

```python
cart = pd.read_csv('cart.csv')
```

```python
cart
```

```python
checkout = pd.read_csv('checkout.csv')
```

```python
checkout 
```

```python
purchase = pd.read_csv('purchase.csv')
```

```python
purchase
```

```python
#Combining visits and cart using a left merge.
visits_cart = pd.merge (visits, cart, how = 'left')
```

```python
visits_cart
```

```python
#How long is the merged DataFrame?
len(visits_cart)
```

```python
#How many of the timestamps are null for the column cart_time?
visits_cart.cart_time.isna().sum()
```

```python
#What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?
not_placing_in_chart = visits_cart.cart_time.isna().sum()*100/len(visits_cart)
not_placing_in_chart
```

```python
#What percentage of users put items in their cart, but did not proceed to checkout?
cart_checkout = pd.merge(cart, checkout, how = 'left')
len(cart_checkout)
cart_checkout.checkout_time.isna().sum()
not_proceed_to_checkout = cart_checkout.checkout_time.isna().sum()*100/len(cart_checkout)
not_proceed_to_checkout
```

```python
#Merging all four steps of the funnel, in order, using a series of left merges
all_data = pd.merge(visits, cart, how = 'left').merge(checkout, how = 'left').merge(purchase, how = 'left')
all_data
```

```python
#What percentage of users proceeded to checkout, but did not purchase a t-shirt?
len(all_data)
all_data.purchase_time.isna().sum()
not_purchase = all_data.purchase_time.isna().sum()*100/len(all_data)
not_purchase
```

```python
#Which step of the funnel is weakest (i.e., has the highest percentage of users not completing it)?
print(not_placing_in_chart)
print(not_proceed_to_checkout)
print(not_purchase)
```
